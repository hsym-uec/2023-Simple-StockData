```python
#初期化
#import
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

service = Service(executable_path=r'/usr/bin/chromedriver')

# ブラウザをheadlessモード実行
options = webdriver.ChromeOptions()
#ヘッドレスモード（バックグラウンドで起動）で実行。コラボの場合、必須。
options.add_argument('--headless')
#サンドボックスモードの解除。これも必須。
options.add_argument('--no-sandbox')
#これも設定した方がよい。
options.add_argument('--disable-dev-shm-usage')
#インスタンス化
#driver = webdriver.Chrome('chromedriver',options=options)
driver = webdriver.Chrome(service=service, options=options)

#指定したドライバーが見つかるまで待機
driver.implicitly_wait(10)


#urlの指定
url="https://www.ssg.ne.jp/login"

driver.get(url)
time.sleep(3)

print("サイトのタイトル：", driver.title)

html = driver.page_source
#BeautifulSoup(html,"lxml")
#login = "xxxxxxxxx"
#password = "xxxxxxxxx"

course_code = "33954"
course_password = "192749"

####⇩自分のチームコードとチームパスワードを入力する
user_code = "8383"
user_password = "394394"

# ユーザー名入力
#driver.find_element(By.XPATH,"//*[@id='contentForm']/form/table/tbody/tr[1]/td/input").send_keys(login)
# driver.find_element(By.XPATH,"//*[@id="course_code"]").send_keys(course_code)

#学校コード
driver.find_element(By.ID,'course_code').send_keys(course_code)

#学校パスワード
# パスワード入力
driver.find_element(By.ID,'course_password').send_keys(course_password)

#チームコード
driver.find_element(By.ID,'user_code').send_keys(user_code)

#チームパスワード
driver.find_element(By.ID,'user_password').send_keys(user_password)



#ログインボタンクリック
btn = driver.find_element(By.XPATH,"/html/body/main/div/div/div/div[2]/form/button")
#btn = driver.find_element(By.XPATH,"//*[@id='contentFormBtn']")
#btn = driver.find_element(By.ID,'button')
btn.click()

time.sleep(5)

#html = driver.page_source
#print("サイトのタイトル：", driver.title)
#BeautifulSoup(html,"lxml")

```

```python
#urlの指定
performance_url="https://www.ssg.ne.jp/performances"
driver.get(performance_url)
time.sleep(3)

#print("サイトのタイトル：", driver.title)
#ranking = driver.find_element(By.XPATH, "/html/body/main/div/div[3]/div[2]/table")
#for i in range(len(ranking)):
#print(ranking.text)

tableElem = driver.find_element(By.CLASS_NAME, "bl_couseDate")
coursedate = tableElem.find_elements(By.TAG_NAME, "p")
print(coursedate[0].text)
print(coursedate[1].text)

tableElem = driver.find_element(By.CLASS_NAME, "Dealings")
trs = tableElem.find_elements(By.TAG_NAME, "tr")
ranking = []
# ヘッダ行は除いて取得
ths = trs[0].find_elements(By.TAG_NAME, "th")
header = []
for j in range(0,len(ths)):
  header.append(ths[j].text)
#print(header)
ranking.append(header)
for i in range(1,len(trs)):
  tds = trs[i].find_elements(By.TAG_NAME, "td")
  line = []
  for j in range(0,len(tds)):
      line.append(tds[j].text)
  ranking.append(line)
#  print(line)
print(ranking)
```

```python
import pandas as pd

member = [['チーム名', '名前'], ['skyriver', '笹川'], ['Sarah', '小原'], ['kgy', '林'], ['shishamo', '出口'], ['bradio', '柳谷'], ['privet', '細川'], ['R15', '吉岡'], ['IFrit', '武田'], ['Yosshy', '東川'], ['Turnip2', '小川'], ['hirotake', '新妻'], ['SoftmaxLove', '名執'], ['LGM', '橋山']] 
df_member = pd.DataFrame(member[1:], columns=member[0])
print(df_member)
df_member_set = df_member.set_index('チーム名')
print(df_member_set)

df_ranking = pd.DataFrame(ranking[1:], columns=ranking[0])
print(df_ranking)
df_ranking_set = df_ranking.set_index('チーム名')
print(df_ranking_set)

df_ranking2 = df_member_set.join(df_ranking_set)
print(df_ranking2)

df_ranking2['資産合計'] = df_ranking2['資産合計'].str.replace(',', '').astype(int)
#print(df_ranking2.sort_values('順位'))

print(df_ranking2.sort_values('資産合計', ascending=False))

```
