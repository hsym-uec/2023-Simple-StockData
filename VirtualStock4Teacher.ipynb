```python
#初期化
#import
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

service = Service(executable_path=r'/usr/bin/chromedriver')

# ブラウザをheadlessモード実行
options = webdriver.ChromeOptions()
#ヘッドレスモード（バックグラウンドで起動）で実行。コラボの場合、必須。
options.add_argument('--headless')
#サンドボックスモードの解除。これも必須。
options.add_argument('--no-sandbox')
#これも設定した方がよい。
options.add_argument('--disable-dev-shm-usage')
#インスタンス化
#driver = webdriver.Chrome('chromedriver',options=options)
driver = webdriver.Chrome(service=service, options=options)

#指定したドライバーが見つかるまで待機
driver.implicitly_wait(10)

```

```python
#ログインまで
#urlの指定
url="https://www.ssg.ne.jp/login"

####⇩自分のチームコードとチームパスワードを入力する

course_code = "33954"
course_password = "192749"

user_code = "8383"
user_password = "394394"

driver.get(url)
time.sleep(3)

#print("サイトのタイトル：", driver.title)

html = driver.page_source
#学校コード
driver.find_element(By.ID,'course_code').send_keys(course_code)
#学校パスワード
driver.find_element(By.ID,'course_password').send_keys(course_password)
#チームコード
driver.find_element(By.ID,'user_code').send_keys(user_code)
#チームパスワード
driver.find_element(By.ID,'user_password').send_keys(user_password)

#ログインボタンクリック
btn = driver.find_element(By.XPATH,"/html/body/main/div/div/div/div[2]/form/button")
#btn = driver.find_element(By.XPATH,"//*[@id='contentFormBtn']")
#btn = driver.find_element(By.ID,'button')
btn.click()

time.sleep(5)

html = driver.page_source
print("サイトのタイトル：", driver.title)
print("Logged in")
```
